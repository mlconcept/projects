```python
df1 = pd.concat([df[['ID','YEAR','CRITICS_POINTS','USER_POINTS','SalesInMillions']],
          pd.get_dummies(df[['CONSOLE','CATEGORY','RATING']])],axis =1)
```

```python
def rmse(x_tst,reg_var):
    from sklearn.metrics import mean_squared_error
    housing_predictions = reg_var.predict(x_tst)
    lin_mse = mean_squared_error(y_test, housing_predictions)
    lin_rmse = np.sqrt(lin_mse)
    return print(lin_rmse)
```
