#### Linear Function
A linear function has one independent variable (x) and one dependent variable (y), and has the following form:

y = f(x) = mx + c

- f(x) = the output (the dependant variable)
- x = the input (the independant variable)
- m = slope = is the coefficient of the independent variable. It gives the rate of change of the dependent variable
- c = intercept = is the value of the dependent variable when x = 0. It is also the point where the diagonal line crosses the vertical axis.

```python
# No need to worry about codes, We'll cover it later
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4])
y = 2*x +10
plt.plot(x,y)
plt.show()
```

### Slope and Intercept
Slope: The slope represents the rate of change of y with respect to x

while the y-intercept is the value of y when x is equal to zero.
